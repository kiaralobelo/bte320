```python
def roi(cost, revenue):
    if not isinstance(cost, (int, float)) or not isinstance(revenue, (int, float)):
        print("Error: inputs must be numeric")
        return
    if cost <= 0:
        print("Error: cost must be greater than 0")
        return
    return (revenue - cost) / cost

cost = float(input("Enter cost: "))
revenue = float(input("Enter revenue: "))
print(roi(cost, revenue))
```

```python
def loan_payment(principal, annual_rate, years):
    if principal <= 0 or annual_rate <= 0 or years <= 0:
        raise ValueError("All inputs must be positive numbers")
    r = annual_rate / 12
    n = years * 12
    payment = (r * principal) / (1 - (1 + r) ** -n)
    return payment

principal = float(input("Enter loan amount: "))
annual_rate = float(input("Enter annual interest rate (e.g. 0.05 for 5%): "))
years = float(input("Enter loan term in years: "))

print(f"Monthly payment: ${loan_payment(principal, annual_rate, years):.2f}")
```

```python
def getValues(n):
  assert isinstance (n, int) and n>0, "n must be a positive integer"
  values = []
  for i in range(n):
    try:
      val = float(input("Enter value: "))
      values.append(val)
    except ValueError:
      print("Warning: Invalid Input. Please enter number")
  return values

def calcAverage(alist):
  assert isinstance(alist, list), "Input must be a list"
  if len(alist)<2:
    raise Exception("List must contain at least two values")
  included = alist[2:]
  return sum(included)/len(included)

try:
  n=int(input("How many values? "))
  values=getValues(n)
  print("Average:",calcAverage(values))
except ValueError:
  print("Warning: Numbers must be integers")
except Exception as e:
  print("Error:",e)
```

```python
#in class assignment
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(f"student {self.__name} has already enrolled in {newCourse}")
```

```python
s = Student("john doe", 1234)
print(s.get_name())
print(s.get_courses())
s.enroll('BTE320')
print(len(s))
print(s.get_courses())
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx=(self.x-other.x)**2
        dy=(self.y-other.y)**2
        return (dx+dy)**0.5

class Rectangle:
    def __init__(self):
        self.p1 = Point(0,0)
        self.p2 = Point(3,0)
        self.p3 = Point(0,3)
        self.p4 = Point(3,3)

        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height
```

```python
class Employee:
  def __init__(self,name,wage,hours):
    self.name=name
    self.__hourly_wage=wage
    self.hours=hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self,newWage):
    if newWage<=0:
      print("Negative wage values not allowed.")
    else:
     self.__hourly_wage=newWage

  def calculate_pay(self):
    return self.__hourly_wage*self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}$/hour"

class Manager(Employee):
  def __init__(self,name,wage,hours,bonus):
    super().__init__(name,wage,hours)
    self.bonus=bonus

  def calculate_pay(self):
    return (1+self.bonus)*self.get_wage()*self.hours

employee=Employee("Bob",50,10)
manager=Manager("Alice",100,15,.2)
print(employee)
print(manager)
print(f"Employee {employee.name} weekly salary is: {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is: {manager.calculate_pay()}$ that includes a {manager.bonus *100}% bonus")
```
