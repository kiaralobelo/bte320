```python
n1 = float(input("enter first value"))
n2 = float(input("enter second value"))
op = input("numerical operation (e.g. +, -, etc.): ")

if op == "+":
  res = n1 + n2
elif op == "-":
  res = n1 - n2
elif op == "*":
  res = n1 * n2
elif op == "/":
  res = n1 / n2
elif op == "//":
  res = n1 // n2
elif op == "**":
  res = n1 ** n2
elif op == "%":
  res = n1 % n2
else:
  print("invalid operation")

print(f"result: {res}")
```

```python
import math

ticketprice = 10
defaultaudience = 20
admoney = 0
print("Ad$$    Profit$$")
while admoney <= 200:
  revenue = (defaultaudience + 2 * round(math.sqrt(admoney))) * ticketprice
  cost = 200 + admoney
  profit = revenue - cost
  print(f"{float(admoney)} \t {float(profit)}")
  admoney += 25
```
