```python
sales = float(input("Sales: $"))

cogs = 0.4*sales
gross_profit= sales-cogs

print(f"Sales: {sales}$")
print(f"cost of goods sold: {cogs}$")
print(f"gross profit: {gross_profit}$")
tax_rate = float(input("tax rate: %"))
tax = gross_profit*(tax_rate/100)
net_income = gross_profit-tax
print(f"taxes: {tax}$")
print(f"net income: {net_income}$")
```

```python
n= float(input("number 1 :"))
n2= float(input("number 2 :"))
add = "+"
divide = "/"
subtract = "-"
multiply = "*"
formula = "n + n2"
if add in formula :
  print("n + n2 = " n+n2)
if divide in formula :
  print("n / n2 = " n/n2)
if subtract in formula :
  print("n - n2 = " n-n2)
if multiply in formula :
  print("n * n2 = " n*n2)
```

```python
n1 = float(input("enter first value"))
n2 = float(input("enter second value"))
op = input("numerical operation (e.g. +, -, etc.): ")

if op == "+":
  res = n1 + n2
elif op == "-":
  res = n1 - n2
elif op == "*":
  res = n1 * n2
elif op == "/":
  res = n1 / n2
elif op == "//":
  res = n1 // n2
elif op == "**":
  res = n1 ** n2
elif op == "%":
  res = n1 % n2
else:
  print("invalid operation")

print(f"result: {res}")
```
